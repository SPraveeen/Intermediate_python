- 4 pillars of oops
  - __Encapsulation__ {putting things altogether like creating class} {spices in kitchen kept in one container for quicker response} {for tea making sugar- tea- all in one specific location {it belongs together}}
  - __Abstraction__ {car contains Accel, Brake, Cluc} {engine which is lot more complicated but it's hidden} {actually even don't know the knowledge of engine we can use the car}
  - __Inheritance__ {all of them inherit lot from parents like finance/ height/ colour} {car - sedan/hatchback/suv - but having common something like passenger seat/tyre} {a way of building heirachical real world relationship in code}
  - __Polymophism__ {4+4=8 // 'abc'+'def'=abcdef} - similar syntax different behaviour {consistent syntax/process leading to a different result depending on the input variables}

```python
class Car:
    def __init__(self,car_name,car_type,car_manufacturer):
        self.name=car_name
        self.type=car_type
        self.manu=car_manufacturer

    def drive(self):
        print('car is driving')
```

```python
c1=Car("alto k10","Hatchback","Suzuki")
```

```python
c1.drive()
```

```python
class Car:
    def __init__(self,car_name,car_type,car_manufacturer):
        self.name=car_name
        self.type=car_type
        self.manu=car_manufacturer

    def drive(self):
        print(f'{self.name} is driving')
```

```python
c1=Car("alto k10","Hatchback","Suzuki")
```

```python
c1.drive()
```
